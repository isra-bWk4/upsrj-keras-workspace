# **Red densa con Keras para predicción lineal**
## Objetivo
Implementar una red neuronal completamente conectada (DNN) simple usando Keras, para predecir la resistencia del concreto(Strength) a partir de sus componentes químicos

1. **Cargar y preparar los datos**

Usaremos el data set de ejemplo UC Irvine. Para ello seguiremos el README.md de [ucimlrepo package](https://github.com/uci-ml-repo/ucimlrepo)

```python
import pandas as pd

#URL del Dataset a utilizar
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/concrete/compressive/Concrete_Data.xls"

# Carga de datos de URL
data = pd.read_excel(url)

#Mostrar las primeras filas
data.head()
```

Ahora es necesario separar los predictores y la variable objetivo de nuestra base de datos. Para ello, sleccionaremos la columna **"Concrete compressive strength(MPa, megapascals)"** como nuestro selector, y el resto como nuestros predictores

```python
strength_column = "Concrete compressive strength(MPa, megapascals) "
```

```python
# Separar predictores
predictors = data.drop(columns=[strength_column])
predictors.head()
```

```python
# Separar selector
selector = data["Concrete compressive strength(MPa, megapascals) "]
selector.head()
```

```python
n_cols = predictors.shape[1]
print(n_cols)
```

2. **Definir el modelo**
Necesitamos generar un modelo secuencial con una red neuronal densa (DNN) con una capa de entrada, una capa oculta y una capa de salida.

La capa de salida contendrá solo una neurona sobre la cual va divergir todo el contenido de la capa previa.
imagennn.jpg

```python
from keras.models import Sequential
from keras.layers import Dense

#Creamos un modelo secuencial
model = Sequential()

#Definimos capa de entrada
model.add(Dense(5, activation='relu', input_shape=(n_cols, )))

#Definimos una capa oculta
model.add(Dense(5, activation='relu'))

#Definimos una capa de salida (Una sola  neurona)
model.add(Dense(1))

#compilamos el modelo
model.compile(optimizer='adam', loss='mean_squared_error')

```

3. **Entrenar y evaluar**
Necesitamos ingresar el contenido del dataset a nuestra red DNN para obtener un resultado.

Posteriormente necestiamos generar una predicción con el modelo alimentado, y evaluar que tan efectivo es nuestro modelo para predecir variables.

```python
# Entrenamiento
history = model.fit(predictors, selector, epochs=50, verbose=1)

#Predicción (con los primeros elementos de los predictores)
predictions = model.predict(predictors)
print(predictions)

```

```python
import matplotlib.pyplot as plt

# Gráfico de perdida
plt.plot(history.history['loss'])
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.grid(True)
plt.show()
```

```python
# Gráfico de dispersión
plt.figure(figsize=(6,6))
plt.scatter(selector, predictions, alpha=0.6)
plt.xlabel('Actual Strength (MPa)')
plt.ylabel('Predicted Strength (MPa)')
plt.title('Predicted vs Actual')
plt.grid(True)
plt.plot([selector.min(), selector.max()], [selector.min(), selector.max()], 'r--')
plt.show()
```

## **Ejercicio: Mejorar una red densa con Keras**

## Objetivo

Predecir la resistencia del concreto mejorando la red original mediante:
- Más capas ocultas
- Más epochs
- Integración de funciones de activación no lineales (sigmoidea)


## Instrucciones

1. Aumenta la densidad de la capa de entrada de 8 a 10 nodos.
2. Modifica las capas ocultas de la red para que sean 3, con el siguiente numero de nodos respectivamente: 10, 8, 6
3. Cambia la función de activación de la capa de salida por una sigmoidea ("sigmoid")
4. Alimenta la red con un valor de 200 epochs.

```python
from keras.models import Sequential
from keras.layers import Dense

#Creamos un modelo secuencial
model = Sequential()

#Definimos capa de entrada
model.add(Dense(10, activation='relu', input_shape=(n_cols, )))

#Definimos una capa oculta
model.add(Dense(10, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(6, activation='relu'))


#Definimos una capa de salida (Una sola  neurona)
model.add(Dense(1, activation='linear'))

#compilamos el modelo
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Entrenamiento
history = model.fit(predictors, selector, epochs=200, verbose=1)

#Predicción (con los primeros elementos de los predictores)
predictions = model.predict(predictors)
print(predictions)
```

```python
import matplotlib.pyplot as plt

# Gráfico de perdida
plt.plot(history.history['loss'])
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.grid(True)
plt.show()
```

```python
# Gráfico de dispersión
plt.figure(figsize=(6,6))
plt.scatter(selector, predictions, alpha=0.6)
plt.xlabel('Actual Strength (MPa)')
plt.ylabel('Predicted Strength (MPa)')
plt.title('Predicted vs Actual')
plt.grid(True)
plt.plot([selector.min(), selector.max()], [selector.min(), selector.max()], 'r--')
plt.show()
```
